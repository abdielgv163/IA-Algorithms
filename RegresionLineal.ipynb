# Modelo de regresión lineal simple y correlación

El **análisis de regresión** es la parte de la estadística que se ocupa de investigar la relación existente entre dos o más variables asociadas en una forma no determinística. 


La relación matemática determinística más simple entre dos variables $X$ y $Y$ es una **relación lineal** del tipo:

$$y = \beta_0 + \beta_1 X$$

Que representa al conjunto de pares $(x,y)$ para los cuales $y=\beta_0 + \beta_1 x$ determina una línea recta con pendiente $\beta_1$ e intersección con el eje $y$ en $\beta_0$.

De modo general vamos a estar trabajando con una variable $x$ que llamaremos **variable independiente, pronosticadora o explicativa**. Esta será aquella a la cuál nosotros le asignamos determinado valor (o es un valor que ya disponemos); y también trabajaremos con una **variable aleatoria** $Y$  junto con su valor observado $y$. Y a esta variable aleatoria $Y$ le llamaremos **variable dependiente o de respuesta**.

De manera que usualmente trabajaremos con pares de datos $(x_i, y_i)$ es decir, pares de datos correspondientes a diferentes valores de la variable independiente y su valor $y_i$ asociado al valor $x_i$. Y a partir de estos pares de datos podremos obtener una **gráfica de dispersión**, permitiéndonos visualizar muy generalmente cuál es el comportamiento de estas variables. Gráfica en la cuál percibiremos cada par $(x_i, y_i)$ como un punto.

### Ejemplo: Gráfica de dispersión

A continuación se presenta una serie de datos recopilados en el artículo **"Analysis of ocular surface area for comfortable VDT workstation layout"**. En dónde se recopilan pares de datos que representan el ancho de la fisura parprebal $\big(x\big)$ y el área de superficie ocular $\big(y\big)$. Esto debido a un análisis respecto a la fatiga e irritación de los ojos a causa de Terminales con pantalla de visualización.

```python
import pandas as pd


# Datos X y Y
x = {"Fisura Parprebal":
    [0.40, 0.42, 0.48, 0.51, 0.57, 0.60, 0.70, 0.75, 0.75, 0.78, 0.84, 0.95, 0.99, 1.03, 1.12, 1.15, 1.20, 1.25, 1.25, 1.28, 1.30, 1.34, 1.37, 1.40, 1.43, 1.46, 1.49, 1.55, 1.58, 1.60]
}

y = {
    "Area de superficie ocular":
    [1.02, 1.21, 0.88, 0.98, 1.52, 1.83, 1.50, 1.80, 1.74, 1.63, 2.00, 2.80, 2.48, 2.47, 3.05, 3.18, 3.76, 3.68, 3.82, 3.21, 4.27, 3.12, 3.99, 3.75, 4.10, 4.48, 3.77, 4.34, 4.21, 4.92]
}

# Los juntamos en un sólo diccionario
data = {**x, **y}

# Creamos un DataFrame a partir de los datos
pd.DataFrame.from_dict(data)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt


# Creamos una figura con 1 solo eje
fig, ax = plt.subplots()

# Graficamos sobre el eje
ax.scatter(x, y, color='purple')
```
